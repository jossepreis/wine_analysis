```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
link = r"C:\Users\josse\OneDrive\Documents\GitHub\checkpoint_4\checkpoint 4\csv\csv_croix.csv"
df_croix = pd.read_csv(link)
```

```python
link = r"C:\Users\josse\OneDrive\Documents\GitHub\checkpoint_4\checkpoint 4\csv\date.csv"
df = pd.read_csv(link)
df_usa_pinot = df[(df['country']=='US')&(df['variety']=='Pinot Noir')]
df_usa_chardonay = df[(df['country']=='US')&(df['variety']=='Chardonnay')&(df['price']<=300)]
```

```python
df_croix['variety'].value_counts()
```

```python
avg_croix = np.mean(df_croix['points'])
avg_pinot_usa = np.mean(df_usa_pinot['points'])
price_avg_pinot_usa = np.mean(df_usa_pinot['price'])
price_min_pinot_usa = np.min(df_usa_pinot['price'])
price_max_pinot_usa = np.max(df_usa_pinot['price'])
```

```python
print(f'note moyenne Domaine des croix : {avg_croix}')
print(f'note moyenne usa pinot noir : {avg_pinot_usa}')
```

```python
print(f'prix moyen pinot noir usa : {price_avg_pinot_usa}')
print(f'prix min pinot noir usa : {price_min_pinot_usa}')
print(f'prix mx pinot noir usa : {price_max_pinot_usa}')
```

```python
fig_pinot = px.scatter(data_frame=df_usa_pinot,x='points',y='price')

fig_pinot.update_layout(template='plotly_dark', title='pinot : price by points' ,height = 500,width=800)
fig_pinot.show()
```

```python
avg_croix = np.mean(df_croix['points'])
avg_chardonay_usa = np.mean(df_usa_chardonay['points'])
price_avg_chardonay_usa = np.mean(df_usa_chardonay['price'])
price_min_chardonay_usa = np.min(df_usa_chardonay['price'])
price_max_chardonay_usa = np.max(df_usa_chardonay['price'])

print(f'note moyenne Domaine des croix : {avg_croix}')
print(f'note moyenne usa chardonay : {avg_chardonay_usa}')
print(f'prix moyen chardonay usa : {price_avg_chardonay_usa}')
print(f'prix min chardonay usa : {price_min_chardonay_usa}')
print(f'prix mx chardonay usa : {price_max_chardonay_usa}')
```

```python
fig_chardonay = px.scatter(data_frame=df_usa_chardonay,x='points',y='price')

fig_chardonay.update_layout(template='plotly_dark', title='chardonnay : price by points' ,height = 500,width=800)
fig_chardonay.show()
```
